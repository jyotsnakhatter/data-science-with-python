```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
numpy.random.seed(7)
```

You can put initialise the random no generator with any seed. I have chosen 7.

#### This analysis is a binary classification problem, where based on a sampled data with features of commodities- trade value in USD, weight in Kg and their quantity of trade, we are finding out if the commodity is exported or imported.

```python

dataset = numpy.loadtxt("C:/Users/parth/datasets/neural_network_less2.txt", delimiter=",")

X = dataset[:,0:3] #The first three columns are the features

Y = dataset[:,3] #binary column where import is 1 and export is 0
Y
```

```python
# create model
model = Sequential()
model.add(Dense(12, input_dim=3, activation='relu'))
#we have developed a fully connected network structure with 3 layers. 
#Because they are fully connected, we are using the dense class

model.add(Dense(8, activation='relu'))
#we have used the rectifier activation function since it is a preferred activation function

model.add(Dense(1, activation='sigmoid')) 
#in the last layer, we are using a sigmoid function so that we get the output between 0 and 1 with our threshold as 0.5
```

We are using the sequential model- which is a stack of layers in a feed forward neural network in Keras.

```python
# Compile model

#since the machine, I am executing on doesn't have a GPU connected, the compilation will be done by TensorFlow CPU

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
#the loss function in our model is a logarithmic loss function, with adam as the gradient descent algorithm.
```

```python
# Fit the model
model.fit(X, Y, epochs=150, batch_size=10)
#bath size is the no of instances to evaluated for a weight update
```

```python
# evaluate the model
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

### Therefore, our feed forward network suggests that using trade_value, weight and quantity to find out if a commodity was imported or exported is only 37% accurate!

We could use more sophesticated neural nets (with better computational power on our systems)!
